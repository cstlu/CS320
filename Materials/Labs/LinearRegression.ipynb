```python
import numpy as np
```

```python
m = 50
X = np.random.randint(10,size = (m,3))
```

```python
Y = np.sum([3,-5,6]*X,axis=1).reshape([m,1])
```

```python
Y
```

```python
X.dot(np.array([3,-5,6]).T)
```

```python
theta = np.random.uniform(-10,10,[3,1])
```

```python
theta
```

```python
def h_function(X,theta):
    return X.dot(theta)
```

```python
def cost_function(X,Y,theta):
    return np.sum((h_function(X,theta) -Y)**2)/(2.0*m)
```

```python
cost_function(X,Y,theta)
```

```python
def Gradient_function(X,Y,theta):
    return (X.T.dot(h_function(X,theta) - Y))/m
```

```python
Gradient_function(X,Y,theta)
```

```python
def GradientDescent(X,Y,theta,epoch,learningrate = 0.003):
    logs =[]
    for _ in range(epoch):
        theta = theta - learningrate*Gradient_function(X,Y,theta)
        loss = cost_function(X,Y,theta)
        print("Loss: ",loss)
        logs.append(loss)
    return (theta,logs)
```

```python
theta,logs = GradientDescent(X,Y,theta,1000)
```
