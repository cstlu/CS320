```python
import numpy as np
import matplotlib.pyplot as plt
```

# Data generator

```python
m = 50
X = np.random.randint(10,size = (m,2))
#X = np.c_[ X, np.ones(m) ]
```

```python
def sigmoid(X,theta):
    return 1./(1+np.exp(-np.dot(X,theta)))
```

```python
Y = sigmoid(X,np.array([[3,-5]]).T)
```

```python
Y = np.array([0 if i < 0.5 else 1 for i in Y]).reshape([m,1])
```

```python
color= ['red' if l == 0 else 'green' for l in Y]
plt.scatter(X[:,0], X[:,1], color=color)
#plt.plot(X,X.dot(np.array([[3,-5]]).T), linestyle='-')
plot_x = X[:,0]
# x1*3 - x2*5 = 0
plot_y = X[:,0]*3/5
plt.plot(plot_x, plot_y);
plt.show()
```

# Traning data

```python
theta = np.random.uniform(-10,10,[2,1])
```

```python
theta
```

```python
def h_function(X,theta):
    return sigmoid(X,theta)
```

```python
def cost_function(X,Y,theta):
    epsilon = 1e-5
    return -np.sum(Y*np.log(h_function(X,theta)+epsilon) + (1 - Y)*np.log(1 - h_function(X,theta) + epsilon))/m
```

```python
cost_function(X,Y,theta)
```

```python
def Gradient_function(X,Y,theta):
    return (X.T.dot(h_function(X,theta) - Y))/m
```

```python
Gradient_function(X,Y,theta)
```

```python
def GradientDescent(X,Y,theta,epoch,learningrate = 0.003):
    logs =[]
    for i in range(epoch):
        theta = theta - learningrate*Gradient_function(X,Y,theta)
        loss = cost_function(X,Y,theta)
        logs.append(loss)
        if i % 50 == 0:
            print("Loss: ",loss)
            # early stopping              
            #if i > 50 and (loss - logs[-2] < 1e-5):
                #print("early stopping")
                #break
            plt.scatter(X[:,0], X[:,1], color=color)
            plot_x = X[:,0]
            plot_y = -X[:,0]*theta[0]/theta[1]
            plt.plot(plot_x, plot_y);
            plt.show()
    return (theta,logs)
```

```python
theta,logs = GradientDescent(X,Y,theta,10000)
```

```python
plt.plot(range(10000),logs)
```
